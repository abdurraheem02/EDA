```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#checking whether any missing values are present in the dataset
data.isnull().sum()
```

```python
data_country=pd.read_excel('/content/Country-Code.xlsx')
data_country.info()
```

```python
#merging two DataFrames data and data_country based on the 'Country Code' column using a left join.
final_data=pd.merge(data,data_country,on='Country Code',how='left')
final_data.head(2)
```

```python
country_names=final_data.Country.value_counts().index
```

```python
country_val=final_data.Country.value_counts().values
```

```python
#creating a pie chart representing the top 3 values in the country_val along with their labels and percentage

plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.savefig("Top 3 Countries.png")
```

Observation:India is the major contributor in the zomato records or transactions after that USA and then UK

```python
final_data.columns
```

```python
# Grouping the data by 'Aggregate rating', 'Rating color', and 'Rating text', count the occurrences, reset the index, and rename the count column as 'Rating Count
ratings=final_data.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Observations:
1.When aggregate rating is between 4.5 to 4.9 then it is "excellent"
2.When aggregate rating is between  4.0to 4.4 then it is "very good"
3.When aggregate rating is between 3.5 to 3.9 then it is "good"
4.When aggregate rating is between 2.5 to 3.4 then it is "average"
5.When aggregate rating is between 1.8 to 2.4 then it is "poor".many people did not rated

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating',y='Rating Count', data=ratings)
plt.savefig("Rating Count vs Aggregate Rating.png")
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color', data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

Observation:
Not rated count is high and maximum number of rating are between 3.0 to 3.8

```python
# Create a count plot of 'Rating color' from the 'ratings' DataFrame using a custom color palette.
sns.countplot(x='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```
